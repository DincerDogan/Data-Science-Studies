# Pandas dataframe

```python
import numpy as np
```

```python
#20 ile 100 arası 20 adet sayı içeren liste elde etmek.
temp = np.random.randint(low = 20, high =100, size = [20,])
#20 adet string oluşturup aşağıdaki kelimeler ile random seçerek liste oluşturma
name = np.random.choice(['Tarik','Alex','Mehmet','Arif'],20)
# 20 adet rassal olarak sayı seçmek.
random = np.random.choice([10,11,13,12,14],20)
```

```python
a = list(zip(temp, name, random))# zip ile obje'ye dönüştürüp list'e cast işlemi.
```

```python
# data frame oluşturma
df = pd.DataFrame(data = a, columns=['temp','name','random']) 
```

```python
df # df ekrana bastırmak.
```

```python
type(df)
```

```python
# dictionary ile data.frame oluşturma.
d1 = pd.DataFrame({
    "city" : ["lucknow","kanpur","agra","delhi"],
    "temperature" : [32,45,30,40]
})
```

```python
d1
```

```python
# Dictionary ile data frame oluşturmak.
d2 = pd.DataFrame({
    "city" : ["delhi","lucknow","kanpur"],
    "humidity" : [68,65,75]
})
```

```python
d2
```

```python
df = pd.merge(d1,d2, on='city') #inner join
```

```python
df
```

```python
pd.merge(d1, d2, on =['city'], how='left') # left join
```
