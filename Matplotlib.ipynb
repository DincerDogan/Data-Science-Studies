# Matplotlib'e Giri≈ü

```python
import numpy as np # numpy import
import pandas as pd # pandas import
import matplotlib.pyplot as plt # matplot import
```

```python
# Plotting two 1-D numpy arrays
x = np.linspace(5, 100, 100) # 5 ve 100 arasƒ± 100 veri
y = np.linspace(10, 1000, 100) # 10 ile 1000 arasƒ± 100 veri

plt.plot(x, y) # ve plot
```

```python
# plotu g√∂stermek.
plt.show()
```

```python
# ayrƒ±ca list vererek de plot √ßizilebilir.
plt.plot([1, 4, 6, 8], [3, 8, 3, 5])
plt.show()
```

```python
# x ve y ekseni olu≈üturma
plt.plot(x, y)

# x ve y ekseninin etiketlerinin yazƒ±lmasƒ±
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's Law") # plot ba≈ülƒ±ƒüƒ±

# x ve y eksenlerini limitlemek.
plt.xlim([20, 80])
plt.ylim([200, 800])
plt.show()
```

```python
# renk ve √ßizgi tipini deƒüi≈ütirmek.

# x ve y dizilerini olu≈üturmak.
x = np.linspace(0, 10, 20)
y = x*2

# √ßizgi rengini mavi ve tipini artƒ± i≈üaretine √ßevirelim.
plt.plot(x, y, 'b+')

# tekrar x ve y eksenine etiketlerini ekleyelim.
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's Law")

plt.show()
```

```python
# birden fazla √ßizgiyi aynƒ± plot'a koymak.
x = np.linspace(0, 5, 10) # dizi olu≈üturma
y = np.linspace(3, 6, 10) # dizi olu≈üturma.

# 3 farklƒ± eƒüriyi √ßizmek: y, y**2 and y**3 
plt.plot(x, y, 'r-', x, y**2, 'b+', x, y**3, 'g^')
plt.show()
```

```python
x = np.linspace(1, 10, 100) # array 1
y = np.log(x) # array 2

# fig√ºr√º initialize etme.
plt.figure(1)

# 1 satƒ±r ve 2 kolonlu plot olu≈üturmak.

# ilk plot'un fig√ºrde √ßizdirilmesi.
plt.subplot(121)     #  plt.subplot(1, 2, 1)'a e≈üittir. 
plt.title("y = log(x)")
plt.plot(x, y)

# ikinci plot'u fig√ºr 1'de √ßizdirmek.
plt.subplot(122)
plt.title("y = log(x)**2")
plt.plot(x, y**2)
plt.show()
```

```python
# √∂rnek data'nƒ±n olu≈üturulmasƒ± 100 adet 1 ve 10 arasƒ±nda.
x = np.linspace(1, 10, 100)

plt.figure(1)

# subplot 1
plt.subplot(2, 2, 1)
plt.title("Linear")
plt.plot(x, x)


# subplot 2 
plt.subplot(2, 2, 2)
plt.title("Cubic")
plt.plot(x, x**3)


# subplot 3
plt.figure(2)

plt.subplot(2, 2, 1)
plt.title("Log")
plt.plot(x, np.log(x))

# subplot 4 
plt.subplot(2, 2, 2)
plt.title("Exponential")
plt.plot(x, x**2)


plt.show()

```

# PieChart

```python
labels = 'T√ºrkiye','ABD','√áin','Almanya'
sizes = [15,35,35,15]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

# explode kƒ±smƒ±ndaki sayƒ±lar grafikteki paylarƒ±n ne kadar 
#uzak g√∂r√ºnebileceƒüine karar verir.
# explode = (0, 0, 0, 0) bu ≈üekilde olduƒüunda hepsi beraber 
# yapƒ±≈üƒ±k g√∂r√ºn√ºr.

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
